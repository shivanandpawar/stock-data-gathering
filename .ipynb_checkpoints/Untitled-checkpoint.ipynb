```python
import yfinance as yf

msft = yf.Ticker("LTIM.NS")

# get all stock info
msft.info
```

```python
fin = msft.quarterly_income_stmt.transpose()
```

```python
hist = msft.history(period="14mo")
```

```python
import pandas as pd
hist["Stock_Year"] = pd.to_datetime(hist.index, format='%y-%m-%d').year
hist["Stock_Month"] = pd.to_datetime(hist.index, format='%y-%m-%d').year
hist["Stock_Date"] = pd.to_datetime(hist.index, format='%y-%m-%d').year
```
