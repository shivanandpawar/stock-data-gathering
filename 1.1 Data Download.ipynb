```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import timedelta
import json
```

```python
def fetch_data(company_name):
    try:        
        msft = yf.Ticker(company_name)
        fin = msft.quarterly_income_stmt.transpose()
        fin["FinancialDate"]=pd.to_datetime(fin.index, format='%y-%m-%d').date
        fin["Year"] = pd.to_datetime(fin.index, format='%y-%m-%d').year
        fin["Month"] = pd.to_datetime(fin.index, format='%y-%m-%d').month

        fin["IsWeekend"] = pd.to_datetime(fin.index, format='%y-%m-%d').day_name().isin(['Saturday', 'Sunday'])
        fin["FinancialDate_Adj"] = np.where(fin['IsWeekend'], fin["FinancialDate"]-timedelta(days=2), fin["FinancialDate"])
        hist = msft.history(period="14mo")
        hist["Stock_Date"] = pd.to_datetime(hist.index, format='%y-%m-%d').date
        final = fin.merge(hist, how='left', left_on='FinancialDate', right_on='Stock_Date')
        final["info_json"]=json.dumps(msft.info)
        return final
    except Exception as e:
        print(e)
        return None
```

```python
import time
import os

stock = pd.read_csv("/data/stock_symbols.csv")

skip_counter = 0

print(skip_counter)

counter = 0

df = fetch_data("LTIM.NS")

for ind in stock.index:
    print(stock['Symbol_New'][ind])
    counter = counter+1
    if counter < skip_counter:
        continue
    df_tmp = fetch_data(stock['Symbol_New'][ind])
    if df_tmp is not None:
        df = pd.concat([df, df_tmp])
    if counter % 5 == 0:
        df.to_csv("/data/raw_stock_indicator_0124.csv", index=False)
    print(counter)
    time.sleep(10) 
    

    df.to_csv("/data/raw_stock_ind.csv", index=False)
```
